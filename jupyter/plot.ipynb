```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../results/20220120T180641output.log')
print(df.to_string())
```

```python
# creating a basic histogram
df['e2e-last'].plot(kind='hist', bins=200, title='Histogram of e2e-last', logy=True)
```

```python
plt.plot(df['e2e-last'])
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1)
fig.set_figheight(5)
fig.set_figwidth(20)

axs.plot(df['e-last'])
axs.grid(linewidth='0.5', color='black')
title = "from function ingress (C) to adt egress (E) latency (avg: {:.2f}s, min: {:.2f}s, max: {:.2f}s, 90th: {:.2f}s)".format(df['e-last'].mean()/1000, df['e-last'].min()/1000, df['e-last'].max()/1000, df['e-last'].quantile(0.9)/1000)
axs.set_title(title, size=14)
```

```python
fig, axs = plt.subplots(nrows=1, ncols=1)
fig.set_figheight(5)
fig.set_figwidth(20)

axs.plot(df['e2e-last'])
axs.grid(linewidth='0.5', color='black')
title = "end-2-end (A-E) latency (avg: {:.2f}s, min: {:.2f}s, max: {:.2f}s, 90th: {:.2f}s)".format(df['e2e-last'].mean()/1000, df['e2e-last'].min()/1000, df['e2e-last'].max()/1000, df['e2e-last'].quantile(0.9)/1000)
axs.set_title(title, size=14)
```
